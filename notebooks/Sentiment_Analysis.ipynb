## **Sentiment Analysis**
Basic sentiment analysis models typically identify whether the overall sentiment of a text is
- positive
- negative
- neutral


## Positive : Happy, Positivity
## Neutral:
## Negative: Anxiety, Depression, Social Anxiety

```python
# Install the required libraries
!pip install transformers torch

# Import the necessary libraries
from transformers import pipeline

# Load a pre-trained sentiment analysis model
sentiment_analyzer = pipeline('sentiment-analysis')

# Example text for analysis
texts = [
    "I love this product! It's absolutely amazing.",
    "This is the worst experience I've ever had.",
    "It's okay, not great but not bad either.",
    "I'm feeling very anxious about this situation.",
]

# Perform sentiment analysis on the texts
results = sentiment_analyzer(texts)

# Print the results
for text, result in zip(texts, results):
    print(f"Text: {text}\nSentiment: {result['label']} (Confidence: {result['score']:.2f})\n")
```

## **Emotional Analysis**

Advanced models can classify text into a range of emotions, such as:
- Happiness
- Sadness
- Anger
- Fear
- Surprise
- Disgust
- Anxiety
- Joy
- Love

**Available Models for Emotion Analysis:**

- ***j-hartmann/emotion-english-distilroberta-base*** - A model fine-tuned for emotion recognition in English text.
- ***bhadresh-savani/bert-base-go-emotion*** - Another popular model trained on the GoEmotions dataset from Google, which includes 27 emotion labels.

```python
from transformers import pipeline

# Load a pre-trained emotion analysis model
emotion_analyzer = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base")

# Example texts for analysis
texts = [
    "I'm so happy today!",
    "I'm feeling very sad and down.",
    "This makes me really angry!",
    "I'm terrified of what might happen.",
    "I love spending time with my family.",
]

# Perform emotion analysis on the texts
results = emotion_analyzer(texts)

# Print the results
for text, result in zip(texts, results):
    print(f"Text: {text}\nEmotion: {result['label']} (Confidence: {result['score']:.2f})\n")
```
